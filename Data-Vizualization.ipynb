## Plotting in Matplotlib

```python
%matplotlib inline
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#set some pandas option
pd.set_option('display.notebook_repr_html',False)
pd.set_option('display.max_columns',20)
pd.set_option('display.max_rows',25)
```

`LETS DO SOME PLOTTING`

```python
plt.plot(np.random.normal(size=100), np.random.normal(size=100),'r*')
```

```python
with mpl.rc_context(rc={'font.family':'serif', 'font.weight':'bold', 'font.size':8}):# setting plot environment
    # declaring a figure to hold subplot
    fig = plt.figure(figsize=(6,3))
    #creating a subplot 
    ax1 = fig.add_subplot(121) # argument 121 is for position:1 with row:2 and cols:1
    ax1.set_xlabel('Some Random Variables')
    ax1.set_ylabel('More Random Variables')
    ax1.set_title('Random stuff')
    plt.plot(np.random.normal(size=100),np.random.normal(size=100),'r.')
    
    # creating another
    ax2 = fig.add_subplot(122)
    ax2.set_xlabel('Samples')
    ax2.set_ylabel('Cummulative Sum')
    ax2.set_title('Normal Distributon')
    plt.hist(np.random.normal(size=100),bins=15)
    plt.tight_layout() # Automatically adds padding and adjust the subplot
    
    # saving figure
    plt.savefig('plot.png',dpi=1500) # dpi : dots/inch in the image
    
```

## Plotting in Pandas

```python
# creating a series a then plotting it
normals = pd.Series(np.random.normal(size=40))
normals.plot(grid=True)
```

```python
normals.cumsum().plot(grid=False)
```

```python
variables = pd.DataFrame({
    'normals':np.random.normal(size=50),
    'gamma':np.random.gamma(1,size=50),
    'poission':np.random.poisson(size=50)
})
variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(subplots=True)
```
